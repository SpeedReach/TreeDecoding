```python
import json
import matplotlib.pyplot as plt

def analyze_memory_and_time(file_paths, ylim=None, labels=None, cutoff=20):
    """
    Analyze and plot memory usage and time metrics from multiple files.
    
    Args:
        file_paths (list): List of paths to the JSONL files
        ylim (tuple): Optional y-axis limits (min, max) for memory plot
        labels (list): Optional labels for each file in the legend
        cutoff (int): Number of steps to cut off from the end
    """
    if not isinstance(file_paths, list):
        file_paths = [file_paths]
    
    if labels is None:
        labels = [f"File {i+1}" for i in range(len(file_paths))]
    
    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    all_time_taken = []
    all_memory_usage = []
    all_time_metrics = []
    
    for i, file_path in enumerate(file_paths):
        # Initialize lists for current file
        time_taken_values = []
        memory_usage_lists = []
        time_metrics_lists = []
        
        # Read the JSONL file
        with open(file_path, 'r') as file:
            for line in file:
                data = json.loads(line)
                time_taken_values.append(data["time_taken"])
                memory_usage = data["memory_usage"][:-cutoff] if len(data["memory_usage"]) > cutoff else data["memory_usage"]
                memory_usage_lists.append(memory_usage)
                time_metric = data["time_metric"]
                normalized_time_metric = [t - time_metric[0] for t in time_metric]
                normalized_time_metric = normalized_time_metric[:-cutoff] if len(normalized_time_metric) > cutoff else normalized_time_metric
                time_metrics_lists.append(normalized_time_metric)
        
        # Calculate and print average time taken
        average_time_taken = sum(time_taken_values) / len(time_taken_values)
        all_time_taken.append(average_time_taken)
        print(f"{labels[i]} - Average time taken: {average_time_taken:.4f} seconds")
        
        # Calculate average memory usage
        sums = []
        counts = []
        for sublist in memory_usage_lists:
            for i, value in enumerate(sublist):
                if i >= len(sums):
                    sums.append(0)
                    counts.append(0)
                sums[i] += value
                counts[i] += 1
        
        sums = sums[:-20]
        counts = counts[:-20]
        
        average_memory_usage = [sums[i] / counts[i] for i in range(len(sums))]
        all_memory_usage.append(average_memory_usage)
        
        # Calculate average time metrics
        time_sums = []
        time_counts = []
        for sublist in time_metrics_lists:
            for i, value in enumerate(sublist):
                if i >= len(time_sums):
                    time_sums.append(0)
                    time_counts.append(0)
                time_sums[i] += value
                time_counts[i] += 1

        time_sums = time_sums[:-20]
        time_counts = time_counts[:-20]
        
        average_time_metrics = [time_sums[i] / time_counts[i] for i in range(len(time_sums))]
        all_time_metrics.append(average_time_metrics)
    
    # Plot memory usage
    for i, avg_memory in enumerate(all_memory_usage):
        ax1.plot(avg_memory, label=labels[i])
    
    if ylim:
        ax1.set_ylim(ylim)
    else:
        ax1.set_ylim(10000, 50000)
    
    ax1.set_title('Average Memory Usage Over Tokens')
    ax1.set_xlabel('Token Count')
    ax1.set_ylabel('Average Memory Usage')
    ax1.grid(True)
    ax1.legend()

    
    # Plot time metrics
    for i, time_metric in enumerate(all_time_metrics):
        ax2.plot(time_metric, label=labels[i])

    
    ax2.set_title('Time Taken at Each Step')
    ax2.set_xlabel('Step')
    ax2.set_ylabel('Time (seconds)')
    ax2.grid(True)
    ax2.legend()
    
    # Adjust layout to prevent overlap
    plt.tight_layout()
    
    # Save the figure
    args = file_paths[0].split("/")
    base_filename = args[-1].split(".")[0]
    plt.savefig(f"comparison_{base_filename}.png")
    plt.show()


# Example usage:
# files = ['path/to/first_file.jsonl', 'path/to/second_file.jsonl']

# labels = ['Original', 'Modified']
# analyze_memory_and_time(files, ylim=(10000, 50000), labels=labels)
```
